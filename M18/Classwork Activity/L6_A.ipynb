```python
# Import pandas library
# Used for data manipulation and handling datasets (like tables)
import pandas as pd

# Import numpy library
# Used for numerical operations and working with arrays/mathematical functions
import numpy as np

# Import matplotlib.pyplot
# Used for data visualization (creating graphs and plots)
import matplotlib.pyplot as plt

# Read the CSV file and load it into a DataFrame
# A dataframe is like a table with rows and columns
# 'petrol_consumption.csv' should be in the same folder as your script
dataset = pd.read_csv('petrol_consumption.csv')
```

```python
# Display the first 5 rows of the dataset
# Helps us quickly understand the structure and sample data
dataset.head()

# Generate statistical summary of numerical columns
# Includes mean, min, max , quartiles, etc.
dataset.describe()

# Selecting independent variables (features)
# These columns will be used to predict petrol consumption
x = dataset[['Petrol_tax', 'Average_income', 'Paved_Highways', 'Population_Driver_licence(%)']]

# Selecting dependent variable (target)
# This is what we want to predict
y = dataset['Petrol_Consumption']

# Import function to split dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Splitting data:
# 80% for training the model and 20% for testing it
# random_state ensures reproducibility (same split everytime)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)

# Import Linear Regression model
from sklearn.linear_model import LinearRegression

# Create an object of LinearRegression class
regressor = LinearRegression()

# Train (fit) the model using training data
# The model learns the relationship between input features and output
regressor.fit(x_train, y_train)

# Create a DataFrame to display coefficients of each feature
# Coefficients tell how much each feature impacts the prediction
coeff_df = pd.DataFrame(regressor.coef_, x.columns, columns = ['Coefficient'])

# Display Coefficients
coeff_df

# Used the trained model to make predictions on test data
y_pred = regressor.predict(x_test)

# Create a comparison DataFrame to check actual vs predicted values
df = pd.DataFrame({'Actual' : y_test, 'Predicted': y_pred})

# Display comparison
df

# Import evaluation metrics to measure model performance
from sklearn import metrics

# Mean Absolute Error (MAE)
# Average of absolute differences between actual and predicted values
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, y_pred))

# Mean Squared Error (MSE)
# Average of squared differences (penalizes larger errors more)
print("Mean Squared Error: ", metrics.mean_squared_error(y_test, y_pred))

# Root Mean Squared Error (RMSE)
# Squared of MSE, gives errorm in same unit as target variable
print("Root Mean Squared Error: ", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
